```python
def cInformacion():
  nEstudiantes=int(input('Diga el número de estudiantes \n'))
  Datos=[]
  for i in range(nEstudiantes):
    Info=[]
    nom=input(f'Diga el nombre del estudiante {i+1} de {nEstudiantes} ')
    Info.append(nom)
    edad=int(input(f'Diga la edad {Info[0]} '))
    Info.append(edad)
    sexo=input(f'Diga el sexo de {Info[0]}; M:Mujer, H:Hombre ').upper()
    while(sexo !='H' and sexo !='M'):
      sexo=input(f'Diga el sexo de {Info[0]}; M:Mujer, H:Hombre ').upper()
    Info.append(sexo)
    print('<<< Seleciona tu programa >>>')
    print('1. Ing Ambiental')
    print('2. Ing Civil')
    print('3. Ing Industrial')
    print('4. Contaduria')
    print('5. Psicologia')
    c=True
    while(c):
      prog=int(input('Programa ? '))
      if(prog > 0 and prog <= 5):
        c=False
    if(prog == 1):
      Info.append('Ing Ambiental')
    elif(prog == 2):
      Info.append('Ing Civil')
    elif(prog == 3):
      Info.append('Ing Industrial')
    elif(prog == 4):
      Info.append('Contaduria')
    else:
      Info.append('Psicologia')
    nAsignaturas=int(input('Diga el numero de asignaturas '))
    Asignaturas=[]
    a=0
    for j in range(nAsignaturas):
      asig=input(f'Diga el nombre de la asignatura {j+1} de {nAsignaturas} ')
      Asignaturas.append(asig)
      nota=float(input(f'Diga la nota de {Asignaturas[a]} '))
      Asignaturas.append(nota)
      credito=int(input(f'Diga los creditos de {Asignaturas[a]} '))
      Asignaturas.append(credito)
      a+=3
    Info.append(Asignaturas)
    Datos.append(Info)
    return Datos
#Función Calcular Promedio
def cPromedio(Datos):
  Promedio=[]
  for i in range(len(Datos)):
    a=0
    sCreditos=0
    NxC=0
    for j in range(len(Datos[i][4])//3):
      NxC += Datos[i][4][1+a] * Datos[i][4][2+a]
      sCreditos +=Datos[i][4][2+a]
      a+=3
  Promedio.append(round(NxC / sCreditos,2))
  return Promedio
```
