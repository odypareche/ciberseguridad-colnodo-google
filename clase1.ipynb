```python
nEstudiantes = int(input('Diga el numero de estudiantes \n'))
Datos = []

for i in range(nEstudiantes):
    Info = []
    nom = input(f'\nNombre del estudiante {i+1} de {nEstudiantes}: ')
    Info.append(nom)

    while True:
        try:
            edad = int(input(f'Diga la edad de {nom}: '))
            Info.append(edad)
            break
        except ValueError:
            print("Error: Por favor, ingrese un número válido para la edad.")


    while True:
        sexo = input(f'Diga el Sexo de {nom} (H: Hombre / M: Mujer): ').upper().strip()
        if sexo in ['H', 'M']:
            Info.append(sexo)
            break
        else:
            print("Entrada no válida. Por favor, use 'H' para Hombre o 'M' para Mujer.")

    print('\n<<< Selecciona tu programa >>>')
    print('1. Ing Ambiental\n2. Ing Civil\n3. Ing Industrial\n4. Contaduria\n5. Psicologia')

    while True:
        try:
            prog = int(input('¿Programa (1-5)? '))
            programas = {1: 'Ing Ambiental', 2: 'Ing Civil', 3: 'Ing Industrial', 4: 'Contaduria', 5: 'Psicologia'}

            if prog in programas:
                Info.append(programas[prog])
                break
            else:
                print("Opción fuera de rango (1-5).")
        except ValueError:
            print("Error: Ingrese el número del programa.")

    Datos.append(Info)

print("\n--- Listado Final ---")
for est in Datos:
    print(est)
```

```python
nEstudiantes = int(input('Número de estudiantes a registrar: '))
Datos_Globales = []

for i in range(nEstudiantes):
    print(f"\n--- Registro del Estudiante {i+1} ---")
    info_estudiante = {} # Usaremos un diccionario para mayor claridad

    nombre = input("Nombre del estudiante: ")
    info_estudiante['Nombre'] = nombre

    nAsignaturas = int(input(f"¿Cuántas asignaturas cursa {nombre}?: "))
    lista_materias = []
    suma_notas = 0
    total_creditos = 0

    # Bucle interno para procesar cada materia
    for j in range(nAsignaturas):
        print(f"  > Materia {j+1}:")
        materia = input("    Nombre de la asignatura: ")
        creditos = int(input(f"    Créditos de {materia}: "))
        nota = float(input(f"    Nota obtenida en {materia}: "))

        # Guardamos los datos de la materia
        lista_materias.append({
            'Asignatura': materia,
            'Creditos': creditos,
            'Nota': nota
        })

        # Acumuladores para el promedio
        suma_notas += nota
        total_creditos += creditos

    # Cálculos finales por alumno
    promedio = suma_notas / nAsignaturas if nAsignaturas > 0 else 0

    # Consolidación de datos del alumno
    info_estudiante['Materias'] = lista_materias
    info_estudiante['Promedio'] = round(promedio, 2)
    info_estudiante['Total_Creditos'] = total_creditos

    Datos_Globales.append(info_estudiante)

# --- Reporte Final ---
print("\n" + "="*40)
print("REPORTE DE RENDIMIENTO ACADÉMICO")
print("="*40)
for est in Datos_Globales:
    print(f"Estudiante: {est['Nombre']}")
    print(f"Promedio: {est['Promedio']} | Créditos Totales: {est['Total_Creditos']}")
    print("-" * 20)
```
